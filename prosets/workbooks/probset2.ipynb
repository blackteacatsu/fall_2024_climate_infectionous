## Question 3

Plot the timeseries of MSC daily mean temperature.

````python
temp_msc_daily = pd.read_csv("../data/data_hw2/data_hw2_p3/temp_MSC_daily.csv")
temp_msc_daily['Datetime'] = pd.to_datetime(temp_msc_daily['Datetime'])
temp_msc_daily.set_index('Datetime', inplace=True)

temp_msc_daily['month'] = temp_msc_daily.index.month
temp_msc_daily['year'] = temp_msc_daily.index.year
temp_msc_daily['day'] = temp_msc_daily.index.day
````

````python
temp_msc_daily
````

````python
px.line(temp_msc_daily, y=temp_msc_daily['Temperature'])
````

Calculate and plot the daily climatology of MSC daily mean temperature using the 2005-2015 data. Explain how you calculated the daily climatology.

````python
daily_mean_teamp_05_to_15 = temp_msc_daily.groupby(['month','day'])['Temperature'].mean().reset_index()
daily_mean_teamp_05_to_15
````

````python
px.line(daily_mean_teamp_05_to_15, y='Temperature', labels={'Temperature':'Temperature in Celsius', 'index': 'Indexed by day of year'})
````
